# Résumé des Commandes Docker (Cycle de Vie)

## 1. Construire (Build)
Construit une image à partir d'un Dockerfile.

    docker build -t mon-app:1.0 .

* `-t` : "Tag" (nom et version) de l'image.
* `.` : Cherche le Dockerfile dans le dossier actuel.

## 2. Lancer (Run)
Démarre un nouveau conteneur à partir d'une image.

    docker run -d -p 8080:80 --name site-web mon-app:1.0

* `-d` : Mode détaché (tourne en arrière-plan).
* `-p` : Publie les ports (Hôte:Conteneur).
* `--name` : Donne un nom lisible au conteneur.
* `--rm` : (Optionnel) Supprime le conteneur automatiquement quand il s'arrête.

## 3. Gérer (Management)
Gérer les conteneurs qui tournent.

    # Lister les conteneurs ACTIFS
    docker ps

    # Lister TOUS les conteneurs (actifs et arrêtés)
    docker ps -a

    # Arrêter un conteneur (proprement)
    docker stop site-web

    # Redémarrer un conteneur arrêté
    docker start site-web

    # Voir les logs (la sortie) d'un conteneur
    docker logs site-web

    # Voir les logs en continu (comme 'tail -f')
    docker logs -f site-web

## 4. Supprimer (Remove / Cleanup)
Nettoyer les conteneurs et les images.

    # Supprimer un conteneur (il doit être arrêté d'abord)
    docker rm site-web

    # Forcer la suppression (même s'il tourne)
    docker rm -f site-web

    # Supprimer une IMAGE (tous les conteneurs l'utilisant doivent être supprimés)
    docker rmi mon-app:1.0

    # Nettoyage automatique (supprime tout ce qui n'est pas utilisé)
    docker system prune
