```python
import pandas as pd
import random
import string
from datetime import datetime, timedelta
import os

os.makedirs("data", exist_ok=True)
csv_path = "data/input.csv"

N = 50

def random_name():
    return ''.join(random.choices(string.ascii_uppercase, k=5))

def random_city():
    return random.choice(["Casablanca", "Rabat", "Marrakech", "Tanger", "Agadir", "Fès"])

def random_category():
    return random.choice(["Electronics", "Clothing", "Books", "Toys", "Food"])

def random_date():
    start = datetime(2024, 1, 1)
    end = datetime(2025, 1, 1)
    return (start + (end - start) * random.random()).strftime("%Y-%m-%d")

data = {
    "id": [i+1 for i in range(N)],
    "name": [random_name() for _ in range(N)],
    "city": [random_city() for _ in range(N)],
    "category": [random_category() for _ in range(N)],
    "price": [round(random.uniform(10, 500), 2) for _ in range(N)],
    "created_at": [random_date() for _ in range(N)]
}

df = pd.DataFrame(data)

df.to_csv(csv_path, index=False, encoding="utf-8")

print(f"✅ CSV généré avec {N} lignes -> {csv_path}")
```
